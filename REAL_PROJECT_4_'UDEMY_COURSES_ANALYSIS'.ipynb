```python
import pandas as pd
```

```python
df = pd.read_csv("7. Udemy Courses.csv")
```

```python
df.head(2)
```

# Q1. What are all different subjects for which Udemy is offering courses.

```python
df.subject.unique()
```

# Q2. Which subject has maximum number of courses.

```python
df.subject.value_counts()
```

# Q3. Show all the courses which are Free of Cost.

```python
df[df.is_paid == False]
```

# Q4.Show all the courses which are Paid.

```python
df[df.is_paid == True]
```

# Q5. Which are Top Selling Course.

```python
df.sort_values('num_subscribers', ascending = False)
```

# Q6. Which are the least selling course.

```python
df.sort_values('num_subscribers', ascending=True)
```

# Q7. Show all courses of Graphics Design where price is below 100.

```python
df[(df.subject == 'Graphic Design') & (df.price < '100')]    ## So there is no course price < 100 after run code.
```

# Q8. List out all the courses that are releated with python

```python
df[df.course_title.str.contains('Python')]
```
