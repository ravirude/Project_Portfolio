```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# Information about Data Set
df.info()
```

```python
# Checking the number of missing value in each column
df.isnull().sum()
```

```python
# distribution of legit transaction & farudulent transaction
df['Class'].value_counts()
```

0 is define for Normal Transaction
1 is define for Fraud Transaction

```python
# Seprating the data for Analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Measurement of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transactions
df.groupby('Class').mean()
```

**Under-Sampling**

Build a sample dataset containing similar distribution of normal transaction and Fradulent Transaction

Number of Fraudulent Transaction -->492

```python
legit_sample = legit.sample(n=492)
```

# Concanating two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
 print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```

```python
print('Accuracy on Testing data : ', test_data_accuracy)
```
