# Netflix Exploratory Data Analysis

Introduction.

This is Exploratory Data Analysis using Python of the "Netflix Movies & TV Shows" dataset. The purpose of this project is to find out and visualize the data's main characteristics and trends using statistical methods and data visualization techniques.

About Data

The Netflix Movies & TV Shows dataset can be found on Kaggle. It contains all TV Shows and Movies metadata available on Netflix. The dataset is updated every month. It contains 8807 records and 12 columns.

Netflix was founded on August 29, 1997, as a mail-based rental business. In January 2007, the company launched a streaming media service, introducing video on demand via the Internet.

Objectives

The purpose of this analysis is to answer the following questions:

1.) Number of movies/tv-shows added to the streaming platform by year.
2.) Which month has the most added movies/tv-shows?
3.) Which day has the most added movies/tv-shows?
4.) How many Movies vs. TV-Shows?
5.) Which year has the most releases Movies/TV-Shows?
6.) Which is the oldest movie/tv-show on streaming?
7.) Which countries produce the most movies/tv-shows?
8.) Cast members with the most content

Oldest Movies/TV-Show on Streaming.

```python
df[df['release_year']== 1925]
```

Top 30 Country producing the most movies/TV-Show.


```python
# New dataset for the country count.
country_count = df.copy()
country_count = pd.concat([country_count, df['country'].str.split(",", expand=True)], axis=1)
country_count = country_count.melt(id_vars=["type","title"], value_vars=range(12), value_name="Country")
country_count = country_count[country_count["Country"].notna()]
country_count["Country"] = country_count["Country"].str.strip()
country_count
```

```python
# Country Unique.
country_count.Country.unique()   ### country should start capital letter i.e(Country)
```

```python
# Country with most number of content Streaming
country_count.Country.value_counts()
```

```python
# Top 30 Cast member with the most Contain.
cast_count = df.copy()
cast_count = pd.concat([cast_count, df['cast'].str.split(",",expand=True)],axis=1)
cast_count = cast_count.melt(id_vars=["type","title"], value_vars=range(44),value_name="Cast_name")
cast_count = cast_count[cast_count["Cast_name"].notna()]
cast_count["Cast_name"] = cast_count["Cast_name"].str.strip()
cast_count
```

```python
cast_count.Cast_name.value_counts()[:30]
```

# Phase-5: Visulazation

Movies Vs TV-Show

```python
px.histogram(df, x='type', color='type',
             title="Movies Vs.TV-Shows",
             color_discrete_sequence=px.colors.sequential.Sunsetdark)
```

```python
px.histogram(df, x= 'type', color= 'type',
             title="Movies vs. TV-Shows",
             color_discrete_sequence= px.colors.sequential.Sunsetdark)
```

Number of Movies/TV-Shows added by Year

```python
px.histogram(df, x=df['date_added'].dt.year, color=df['type'],
             title="Netflix number of Movies/TV Show by year",
             color_discrete_sequence=px.colors.sequential.Sunsetdark,
             labels=dict(x="Year", color="Type")
             )
```

# Number of Movies/TV-Shows added by Month

Dropping 'NA' records from the column 'date_added

Dropping 10 records from the column 'date_added' that contain 'NA' values

```python
# Counting the number of 'NA' on the column 'date_added'
df['date_added'].isna().sum()
```

```python
# Dropping 'NA'
df = df.dropna(subset=['date_added'])
```

```python
px.histogram(df, x=df['date_added'].dt.month, color=df
['type'],
                   color_discrete_sequence=px.colors.sequential.Sunsetdark,
                   title="Movies/TV Shows added by Month",
                   labels=dict(x="Month"))
```

Number of Movies/TV-Shows added by Day

```python
px.histogram(df, x= df['date_added'].dt.day,color= df['type'],
             color_discrete_sequence = px.colors.sequential.Sunsetdark,
             title="Movies/TV Shows added by day",
             labels=dict(x="day"))
```

Number of Movies/TV-Shows by year of release

```python
px.histogram(df, x= 'release_year', color='type',
             title="Number of Movies/TV-Shows by year of release",
             color_discrete_sequence= px.colors.sequential.Sunsetdark,
             labels={'release_year':'Year of release'}
                   )
```

```python
px.histogram(df, x= 'release_year', color='type',
             title="Number of Movies/TV-Shows by year of release",
             color_discrete_sequence= px.colors.sequential.Sunsetdark,  
             labels={'release_year':'Year of release'}                     
                   )
```

Top 30 Countries with the most streaming content

```python
px.histogram(country_count, x= 'Country', color= 'type',
        title="Top 30 Countries with the most streaming content",
        color_discrete_sequence= px.colors.sequential.Sunsetdark).update_xaxes(
        categoryorder="total descending",range=(0, 30))
```

Top 30 Cast members with the most titles

```python
px.histogram(cast_count, x= 'Cast_name', color= 'type',
    title="Top 30 Cast members with the most streaming content",
    color_discrete_sequence= px.colors.sequential.Sunsetdark).update_xaxes(
    categoryorder="total descending",range=(0, 30))
```

# Findings

About Netflix

1. There are more Movies than TV-Shows available on streaming. 6131 movies and 2676 tv-shows.

2. 2019 is the year with the most content addition on the streaming platform, 2016 movie/tv-shows added, followed by 2020 with 1879, and 2018 with 1649 total

3. July and December are the months with the most content addition, 827 and 813 movie/tv-shows added.

4. Netflix adds content on the first day of the month more than any other day.

About the content available

1. Among the contents available 1147 of them were originally released in 2018 followed by 2017 with 1032, and 2019 with 1030 total.

2. Pioneers: First Women Filmmakers is the oldest content available on streaming. It's a collection of restored films dating from 1925.

3. The United States is the country that produces the most of the content with 3690 titles, followed by India 1046 titles and the United Kingdom 806 titles.

4. Anupam Kher is the actor with the higher number of titles, 43 films. Anupam Kher is an Indian actor, director, and producer that has appeared in over 500 films.
